```python
import Num_SH as NSH
import numpy as np
import matplotlib.pyplot as plt
```

```python
#folder = "ThreeEqual"
#filename = "three_equal"
folder = "Tau"
filename = "mostly_tau"
#folder = "AllTau"
#filename = "all_tau"
npz = np.load("{}/{}-results.npz".format(folder,filename),allow_pickle=True)
```

```python
npz.files
```

```python
a = npz['results']
```

```python
plt.figure()
plt.plot(a[:,0],a[:,1],'o')
plt.axhline(60)
```

```python
plt.figure()
for i in range(len(a)):
    if a[i,1] > 60:
        plt.plot(a[i,0]/0.1188,a[i,1],'o',color='k')
    else:
        plt.plot(a[i,0]/0.1188, a[i,1],'o', color='0.75')
plt.axhline(60, color='k', linestyle='--')
plt.xlabel(r'$f = \Omega_s h^2 / (\Omega_{\rm DM} h^2)$')
plt.ylabel('Number of Subhalos')
```

```python
ii = npz['index']
mm = npz['mixangle']
ll = npz['L0']
```

```python
M = np.zeros((len(mm), len(ll)))
S = np.zeros_like(M)
for i in range(len(ii)):
    M[ii[i][0], ii[i][1]] = a[i,0] / 0.1188
    S[ii[i][0], ii[i][1]] = a[i,1]
```

```python
plt.figure()
plt.contourf(ll*1e3, mm, M, levels=[0,0.2, 0.4, 0.6, 0.8, 1])
plt.colorbar(label=r"$f = \Omega_s h^2 / (\Omega_{DM} h^2)$")
#cs =plt.contour(L0*1e3, mixangle, NSH, levels=[30,60,90], colors=['w','w','w'], linewidths=[3,5,3], linestyles=[':','-', '--'])
#plt.clabel(cs, inline=1, fmt='%d', fontsize=12)
#plt.xscale('log')
plt.yscale('log')
plt.xlabel(r"Lepton Number $(\times 10^{-3})$")
plt.ylabel("Total Mixing Angle")
plt.text(4.6,2.2e-9,r"$\Omega_s h^2 > \Omega_{DM} h^2$")

plt.xticks([1,3,5])
plt.xlim([1,6])

image_format = 'png'
image_name = '{}/{}-contourplot.png'.format(folder,filename)

plt.savefig(image_name, format=image_format, dpi=1200)
```

```python
plt.figure()
plt.contourf(ll*1e3, mm, M, levels=[0,0.2, 0.4, 0.6, 0.8, 1])
plt.colorbar(label=r"$f = \Omega_s h^2 / (\Omega_{DM} h^2)$")
cs =plt.contour(ll*1e3, mm, S, levels=[30,60,90,120], colors=['w','w','w','w'], linewidths=[3,5,3,3], linestyles=['-.','-', '--',':'])
plt.clabel(cs, inline=1, fmt='%d', fontsize=12)
#plt.xscale('log')
plt.yscale('log')
plt.xlabel(r"Lepton Number $(\times 10^{-3})$")
plt.ylabel("Total Mixing Angle")
plt.text(4.6,2.2e-9,r"$\Omega_s h^2 > \Omega_{DM} h^2$")

plt.xticks([1,3,5,7])
plt.xlim([1,6])

image_format = 'png'
image_name = '{}/{}-contourplot2.png'.format(folder,filename)

plt.savefig(image_name, format=image_format, dpi=1200)
```

# Spectra & Lepton Evo

- electron

```python
data_e = np.load('Neutrino Data/electronx0.00076x1e-10.npz')

plt.figure()
plt.plot(data_e['epsilon'],data_e['epsilon']**2 * data_e['final_distribution'], linestyle = '--')
plt.plot(data_e['epsilon'],data_e['epsilon']**2 * data_e['sterile_evo'][-1], linestyle = ':')
plt.plot(data_e['epsilon'],data_e['epsilon']**2 * data_e['anti_sterile_evo'][-1], linestyle = '-.')
plt.legend(['$f_{\epsilon} +\overline{f_{\epsilon}}$ ', '$f_{\epsilon}$', '$\overline{f_{\epsilon}}$'])

plt.xlabel(r"$\epsilon = p_\nu / T_{cm}$")
plt.ylabel(r"$\nu_s$ distribution")
image_format = 'png' # e.g .png, .svg, etc.
image_name = 'spectra_e.png'

plt.savefig(image_name, format=image_format, dpi=1200)


plt.figure()
plt.semilogx(data_e['T'], 10**4*data_e['lep_evo'])
plt.semilogx(data_e['T'], 10**4*data_e['initial_electron']/(data_e['a']**3*data_e['T']**3), linestyle = '--' )
plt.legend([r'$L_{\nu_e}$ Evolution', r'$L_{\nu_{\mu}}$, $L_{\nu_{\tau}}$ Evolution'])
plt.xlim(2000,10)
plt.xlabel("T (MeV)")
plt.ylabel(r"Lepton Number $(\times 10^{-4})}$")
image_format = 'png' # e.g .png, .svg, etc.
image_name = 'leptonplot_e.png'

plt.savefig(image_name, format=image_format, dpi=1200)
```

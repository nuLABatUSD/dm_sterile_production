```python
import matplotlib.pyplot as plt
import numpy as np
import ODEsolver as solve
from der import trapezoid
```

```python
N = 1000
f_initial = np.zeros(N)
y_0 = np.zeros(N+3)
p = np.zeros(N+7)

y_0[:N] = f_initial
```

$L_{e0} = 7.6 * 10^{-4}$

```python
%%time
y_0[-3:] = [7.6e-4, 0, 1/2000]
#y_0[-2:] = [0, 1/2000]
p[:N] = np.linspace(10/N, 10, N)
p[-7:] = [61.75, 61.75, 7.6e-4, 7.6e-4, y_0[-3], 10**(-10), 0.064]
#p[-7:] = [61.75, 61.75, 7.6e-4, 7.6e-4,7.6e-4, 10**(-10), 0.064]

end, x, y, dx = solve.ODEsolver(1/2000, y_0, 1/2000000, p, 10000, 100, 1/10)

plt.plot(p[:N], p[:N]**2*y[-1,:N],':') 
plt.xlabel('$\epsilon$')
plt.ylabel('$\epsilon^{2}f(\epsilon)$') 
```

```python
plt.figure()
plt.plot(1/x, y[:,-3])
plt.xlim(max(1/x), min(1/x))
plt.xlabel("T")
plt.ylabel(r"$L_{\nu_e}$")
```

# Energy Density of Dark Matter, $\Omega_sh^2$:

```python
m_pc = 1.22e22
T_cmb = 2.369e-10
T_cm03 = (10.75/p[-6])*(4/11)*T_cmb**3
c = 8*np.pi/(3*m_pc**2)*((1/2.13e-39)**2)*p[-1]/(2*np.pi**2)*T_cm03

c*trapezoid(p[:N], p[:N]**2*y[-1,:N])
```
